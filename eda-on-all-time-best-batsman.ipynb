```python
#importing necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#importing dataset
ipl=pd.read_csv("/kaggle/input/ipl-all-time-best-batsman/All-Time-Best-Batsman.csv")
```

```python
#first few rows of dataset
ipl.head()
```

```python
#no of rows and columns in dataset
ipl.shape
```

```python
#checking for missing values
ipl.isnull().sum()
```

```python
#info about each columns name and data type
ipl.info()
```

```python
#perform basic statistical operations on numerical columns
ipl.describe()
```

```python
#extracting top 10 highest ipl run scrorer
top10=ipl.sort_values(by='Runs',ascending=False).head(10)
```

```python
#plotting bar diagram of top 10 highest ipl run scorer
plt.figure(figsize=(16, 12))
plt.barh(top10["Player"], top10["Runs"],color='orange')
plt.xlabel("Runs")
plt.ylabel("Players Name")
plt.title("Top 10 Runs Scorer in IPL")
plt.gca().invert_yaxis()
plt.show()
```

```python
#finding out which teams has most player in top 10 run scorer 
top10['Team'].value_counts()
```

```python
#extracting player who remained most Not Outs in Ipl
MostNO=ipl.sort_values(by='NO',ascending=False).head(10)
MostNO
```

```python
#plotting a bar plot of player who remained most not outs in IPL
plt.figure(figsize=(18, 10))
plt.bar(MostNO["Player"], MostNO["NO"],color='purple')
plt.xlabel("Players Name")
plt.title("Player who remain Most Not Out")
plt.show()
```

```python
#player with least average in IPL
avgl=ipl.sort_values(by='Avg',ascending=True).head(10)
avgl
```

```python
#player with highest average in IPL
avgh=ipl.sort_values(by='Avg',ascending=False).head(10)
avgh
```

```python
#plotting garph of players having highest average
plt.figure(figsize=(14,8))
plt.bar(avgh['Player'],avgh['Avg'])
plt.title('Player with highest average')
plt.xlabel('Players')
```

```python
#player with most 6s in IPL
Sixs=ipl.sort_values(by='6s',ascending=False).head(10)
Sixs
```

```python
#plotting player who has hit most 6s
plt.figure(figsize=(16,12))
plt.barh(Sixs['Player'],Sixs['6s'],color='pink')
plt.xlabel('Sixes')
plt.ylabel("Player's Name")
plt.title('Players with most 6s')
plt.gca().invert_yaxis()
```

```python
#player with most 50s
most50s=ipl.sort_values(by='50',ascending=False).head(10)
most50s
```

```python
#plotting player with most 50s
plt.figure(figsize=(16,12))
plt.bar(most50s['Player'],most50s['50'])
plt.xlabel('Players Name')
plt.ylabel("50s")
plt.title('Players with most 50s')
plt.show
```
